```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
master = pd.read_csv('masterDrugs.csv')
master = master.dropna()
```

```python
##divide master randomly into train and test
choose = np.random.rand(len(master)) < 0.8
train = master[choose]
test = master[~choose]
```

```python
#train model
variables = train.drop(columns=['SUBJECT_ID', 'delirium'])
output = train['delirium']
model = LinearRegression().fit(variables, output)
```

```python
#use model
test_var = test.drop(columns=['SUBJECT_ID', 'delirium'])
predicted = model.predict(test_var)
test['predicted delirium'] = predicted
test
```

```python
coeff = variables.T
coeff['coefficients'] = model.coef_
coeff = pd.DataFrame(coeff['coefficients'])
coeff
```

```python
##evaluate model
R_squared = model.score(variables, output)
print("R squared coefficient: " + str(R_squared))

print('Mean Absolute Error:', metrics.mean_absolute_error(test['delirium'], test['predicted delirium']))
print('Mean Squared Error:', metrics.mean_squared_error(test['delirium'], test['predicted delirium']))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(test['delirium'], test['predicted delirium'])))
```
